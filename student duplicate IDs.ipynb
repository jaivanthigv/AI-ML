```python
def find_duplicates(arr):
    seen = set()
    result = []
    for id in arr:
        if id in seen:
            result.append(True)
        else:
            seen.add(id)
            result.append(False)
    return np.array(result)
T = int(input())
for _ in range(T):
    # Read the IDs for this test case
    arr = np.array(list(map(int, input().split())))
    output = find_duplicates(arr)
    print(output)
```
