```python
#coloumn tarnsfermation
from sklearn.impute import SimpleImputer
```

```python
##Neumerical processing pipeline
import numpy as np
neumeric_processor = Pipeline(
    steps = [("imputation_mean", SimpleImputer(missing_values=np.nan, strategy="mean")),
            ("scaler", StandardScaler())]

)
```

```python
neumeric_processor
```

```python
##Categorical processing 
from sklearn.preprocessing import OneHotEncoder
categorical_processing = Pipeline(
    steps = [("imputation_constant", SimpleImputer(fill_value = "missing", strategy = "constant")),
              ("onehot", OneHotEncoder(handle_unknown="ignore"))]

)
```

```python
categorical_processing
```

```python
#combine processing techniques
from sklearn.compose import ColumnTransformer


preprocessor = ColumnTransformer(

    [("categorical", categorical_processing, ["gender", "City"]),
     ("neumerical", neumeric_processor, ["age", "height"])]
)
```

```python
preprocessor
```

```python
from sklearn.pipeline import make_pipeline
```

```python
pipe = make_pipeline(preprocessor, LogisticRegression())
```

```python
pipe
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.pipeline import Pipeline
from sklearn.ensemble import RandomForestRegressor
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.compose import ColumnTransformer
```

```python
#step 1 preprossor
num_features = ["age", "income"]
cat_features=["gender","city"]
```

```python
num_transformer = StandardScaler()
cat_transformer = OneHotEncoder(handle_unknown = "ignore")
preprocessor = ColumnTransformer([
    ("num", num_transformer, num_features),
    ("cat", cat_transformer,cat_features)
])
```

```python
pipe = Pipeline([
    ("preprocessor", preprocessor),
    ("regressor", RandomForestRegressor())
])
```

```python
param_grid = {
'regressor__n_estimators': [50, 100, 200],
'regressor__max_depth': [None, 10, 20],
'regressor__min_samples_split': [2, 5],
}
```

```python
grid_search = GridSearchCV(pipe, param_grid, cv=5, scoring='r2', n_jobs=-1)
grid_search.fit(X_train, y_train)
```
