```python
import pandas as pd
import numpy as np
```

```python
column_names = ["user_id", "item_id", "rating", "timestamp"]
df = pd.read_csv('u.data', sep="\t",  names = column_names)
```

```python
#Lets get movie titles
movie_titles= pd.read_csv("Movie_Id_Titles.unknown")
movie_titles
```
